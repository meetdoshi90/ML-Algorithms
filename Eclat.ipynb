```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r'C:\Users\meetd\OneDrive\Desktop\Datasets\market-basket-optimisation\Market_Basket_Optimisation.csv')
transactions = []
for i in range(0,75):
    transactions.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence =  0.2, min_lift = 3,  min_length=2, max_length=2)
```

```python
results = list(rules)
print(rules)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
```

```python
resultsinDF = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

```python
resultsinDF.nlargest(n=10, columns = 'Support')
```
